```python
import matplotlib.pyplot as plt
import numpy as np

# Data
epochs = list(range(1, 13))  # Epoch numbers
train_loss = [0.1087, 0.0175, 0.0167,0.0004, 0.0001, 0.0001,0.0227, 0.0059, 0.0001,0.000,0.0001,0.0004]
val_loss = [0.2686, 0.1560, 0.2336, 0.3266, 0.2772, 0.1897, 0.2994, 0.3771, 0.1910, 0.2262,0.214, 0.204]
val_accuracy = [91.36, 94.81, 92.59, 92.35, 94.07, 95.80, 93.70, 93.83, 96.54, 95.93, 93.56,95.32]


# Create the figure and axis
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot Train Loss and Validation Loss
ax1.plot(epochs, train_loss, label="Train Loss", color="blue", marker="o", markersize=5, linestyle="-")
ax1.plot(epochs, val_loss, label="Validation Loss", color="orange", marker="s", markersize=5, linestyle="-")
ax1.set_xlabel("Epochs", fontsize=12)
ax1.set_ylabel("Loss", fontsize=12, color="black")
ax1.tick_params(axis="y", labelsize=10)

# Set custom y-ticks for Loss (half the current major ticks)
major_loss_ticks = np.arange(0, 1.1, 0.1)  # Major ticks at intervals of 0.1
minor_loss_ticks = np.arange(0, 1.1, 0.1 / 2)  # Minor ticks at intervals of 0.05
ax1.set_yticks(major_loss_ticks)
ax1.set_yticks(minor_loss_ticks, minor=True)

# Set Loss Y-Axis range
def fivepercent(x):
    return 0.05*x
ax1.set_ylim(0 - fivepercent(1), 1 + fivepercent(1))

# Add extra horizontal gridlines
ax1.grid(which="major", color="gray", linestyle="-", linewidth=0.7, alpha=0.7)  # Major gridlines
ax1.grid(which="minor", color="gray", linestyle="--", linewidth=0.5, alpha=0.5)  # Minor gridlines

# Second y-axis for Validation Accuracy
ax2 = ax1.twinx()
ax2.plot(epochs, val_accuracy, label="Validation Accuracy", color="green", marker="^", markersize=5, linestyle="-")
ax2.set_ylabel("Validation Accuracy (%)", fontsize=12, color="green")
ax2.tick_params(axis="y", labelsize=10, colors="green")

# Set custom y-ticks for Accuracy (half the current major ticks)
major_acc_ticks = np.arange(0, 101, 10)  # Major ticks at intervals of 10
minor_acc_ticks = np.arange(0, 101, 10 / 2)  # Minor ticks at intervals of 5
ax2.set_yticks(major_acc_ticks)
ax2.set_yticks(minor_acc_ticks, minor=True)

# Set Accuracy Y-Axis range
ax2.set_ylim(0 - fivepercent(100), 100 + fivepercent(100))

# Add title and legend
plt.title("Training and Validation Metrics Over Epochs (volo d5 with LLC_FASD dataset)", fontsize=14, pad=15)
fig.legend(loc="upper center", bbox_to_anchor=(0.5, 1.1), ncol=3, fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```
